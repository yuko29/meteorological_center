```python
import itertools

CONFIG_SCHEMA = [{
        "collection_name": "earthquake",
        "schema":{
            "time": "datetime", 
            "focal_dep": "float", 
            "longitude": "float",
            "latitude": "float",
            "magnitude": "list"
        }
    },
    {
        "collection_name":"reservoir",
        "schema":{
            "time": "datetime",
            "water_supply": "float",
            "percentage": "float",
            "name":"str"
        }
    },
    {
        "collection_name":"electricity",
        "schema":{
            "region":"str",
            "power_usage":"float",
            "power_generate":"float",
            "time":"datetime"
        }
    }]

def get_combinations(dictionary):
    keys = dictionary.keys()
    values = dictionary.values()
    combinations = list(itertools.product(*values))
    results = [{k: v for k, v in zip(keys, combination)} for combination in combinations]
    return results
```

```python
import itertools

data = INSERT_ISP

base_choices = {
    "float": 32.0,
    "datetime": "2023-05-23 00:01:23",
    "str": "a string",
    "list": [{"factory": "南", "magnitude": 0.003}]
}

keys = data.keys()
combinations = list(itertools.product(*(data[key] for key in keys)))

test_cases = []
for combo in combinations:
    count = 0
    for key, value in zip(keys, combo):
        if base_choices[key] != value:
            count = count+1
    if count>1:
        continue
    else:
        test_case = {key: value for key, value in zip(base_choices, combo)}
        test_cases.append(test_case)

# Print the test cases
for i, test_case in enumerate(test_cases):
    print(f"Test Case {i+1}: {test_case}")
```
